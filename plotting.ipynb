```python
import math
import matplotlib.pyplot as plt
import numpy as np
import  pandas as pd
```

```python
x_1 = np.linspace(0,5, 10)
y_1 = x_1 **2
plt.plot(x_1, y_1)
plt.title('Days Squared Chart')
plt.xlabel('days')
plt.ylabel('days squared')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'r')
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'g')
```

```python
fig = plt.figure(figsize=(5, 4), dpi=100)
axes_1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1, y_1, label = 'x/x^2')

axes_1.plot(y_1, x_1, label = 'x/x^2')
axes_1.legend(loc='best')

axes_2 = fig.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(0, 40, 'Message')
```

```python
fig_2, axes_2 = plt.subplots(figsize =(8,4), nrows = 1, ncols =3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1, y_1, 'c')

```

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x_1, y_1, color='navy', alpha = .75, lw=2, ls='-.', marker ='o', markersize = 7, markerfacecolor ='y',
            markeredgecolor ='y', markeredgewidth=4)
# plt.xkcd()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
axes.grid(True, color = '0.6', dashes=(5,2,1,2))
axes.set_facecolor ('#FAEBD7')

fig.savefig('3rd_plot.png')
```
